```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("marketing_campaign1.xlsx")
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors='coerce')
df["Age"] = 2025 - df["Year_Birth"]
df["Children"] = df["Kidhome"] + df["Teenhome"]
product_cols = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts",
                "MntSweetProducts", "MntGoldProds"]
df["TotalSpend"] = df[product_cols].sum(axis=1)
df.drop(columns=["ID", "Year_Birth", "Z_CostContact", "Z_Revenue"], inplace=True)
```

```python
df["Income"].fillna(df["Income"].median(), inplace=True)
```

```python
print("Data Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
print("\nStatistical Summary:\n", df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Income"], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("cleaned_customer_data.csv", index=False)
print("Cleaned data saved as 'cleaned_customer_data.csv'")
```
